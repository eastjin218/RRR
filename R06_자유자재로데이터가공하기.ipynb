## 다수 조건문, &


```r
# 1반 이면서 수학 점수가 50점 이상인 경우
exam %>% filter(class == 1 & math >= 50)
```

```r
# 2반 이면서 영어점수가 80점 이상인 경우
exam %>% filter(class == 2 & english >= 80)
```

```r
# 수학 점수가 90점 이상이거나 영어점수가 90점 이상인 경우
exam %>% filter(math >= 90 | english >= 90)
```

# 집단별 요약하기
- sql에성 where에 having..?

```r
exam %>% 
 group_by(class) %>% 
 summarise( mean_math = mean(math) # math 평균
           , sum_math = sum(math) # math 합계
           , medium_math = median(math) # 중앙값
           , n = n() # count!, 학생수 
          )
```

## 자주 사용하는 요약통계량
- mean
- sd() : 표준편차
- n() : 빈도 
- median() : 중앙값.

```r
midwest <- as.data.frame(ggplot2::midwest)
midwest
```

```r
str(midwest)
```

```r
# 전체 인구대비 미성년 인구
# poptotal L 전체 / popadults는 미성년 인구
midwest  <-  midwest %>% 
 mutate(rate_notadults = ((poptotal -  popadults )/ poptotal) * 100 )
midwest
```

```r
midwest %>% 
 arrange(desc(rate_notadults), by=county)
```

```r
# 분류표
# 미성년 비율등급 
# large = 40dltkd
# middle= 30 - 40
# small = 30 미만
midwest <- midwest %>% 
            mutate(rank = ifelse( rate_notadults >= 40, 'large'
                                 , ifelse( rate_notadults < 40 & rate_notadults >= 30 ,'middle', 'small' ) 
                                ) 
                  )
```

```r
str(midwest)
```

```r
midwest %>%
 group_by(rank) %>% 
 summarise( rank_count =  n())
```

```r
midwest %>%
 select(state, county, poptotal, popasian) %>% 
 mutate( rate_asian = ( popasian / poptotal) *100 ) %>% 
 head(10)
```
